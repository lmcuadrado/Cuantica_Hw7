```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#
min_rho = 0.0
max_rho = 20.0
drho=0.01
N= int((max_rho - min_rho)/drho) #cantidad de puntos o datos que se tomaran
```

```python
def ddR(l,eps,rho,R,dR):
    return -float(2*dR/rho) + float((l*(l+1)/(rho**2))-(2/rho)-e)*R

#def prob(r, R):
#    R = R*R*r*r
 #   R = R/(np.trapz(R,r))
  #  return R
```

```python
def sol(l,eps,drho,N):
    rho = np.zeros(N)
    R = np.zeros(N)
    dR = np.zeros(N)
    
    if l==0:
        R0=1
        dR0=-1
    else:
        R0=0
        dR0=1
   
    rho0=0.0000000001
    R=np.append(R,R0)
    dR=np.append(dR,dR0)
    rho=np.append(rho,rho0)
    
    for i in range (1,N):
        rho0 += drho 
        R0 += drho * dR0
        dR0 += dR0 + ddR(l,eps,rho0,R0,dR0)*drho
    
        R= np.append(R,R0)
        dR= np.append(dR,dR0)
        rho= np.append(rho,rho0)
        
    return rho, R
#print (sol(0,-0.999,0.01,20))
```
