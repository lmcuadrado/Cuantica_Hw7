```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#
min_rho = 0.0
max_rho = 20.0
drho=0.01
N= int((max_rho - min_rho)/drho) #cantidad de puntos o datos que se tomaran
```

```python
def ddR(l,eps,rho,R,dR):
    return -float(2*dR/rho) + float((l*(l+1)/(rho**2))-(2/rho)-eps)*R

#def prob(r, R):
#    R = R*R*r*r
 #   R = R/(np.trapz(R,r))
  #  return R
```

```python
def sol(l,eps,drho,N):
    rho = np.zeros(N)
    R = np.zeros(N)
    dR = np.zeros(N)
    
    if l==0:
        R[0]=1
        dR[0]=-1
    else:
        R[0]=0
        dR[0]=1
   
    rho[0]=0.000000001

    print(rho,R, dR)
    
    for i in range (1,N):
        rho[i] = rho[i-1]+drho 
        R[i]= drho * dR[i-1]
        dR[i] = dR[i-1] + ddR(l,eps,rho[i-1],R[i-1],dR[i-1])*drho
        
    return rho, R


sol(0,-0.999,0.01,20)
```

```python
e = [-0.990099,-0.249,-0.107,-0.2502, -0.1101,-0.111]
for i in range(0,len(e)):
    if i < 3:
        x,y = sol(0,e[i],drho,N)
        x,y = np.array(x),np.array(y)
        y = y/np.sqrt(np.trapz(y**2,x)) 
        plt.plot(x,y)
        plt.axis([0,0.23,-2,2])
    if i >= 3 and i < 5:
        x,y = sol(1,e[i],drho,N)
        x,y = np.array(x),np.array(y)
        y = y/np.sqrt(np.trapz(y**2,x)) 
        plt.plot(x,y)
        plt.axis([0,0.23,-2,2])
    if i==5:
        x,y = sol(2,e[i],drho,N)
        x,y = np.array(x),np.array(y)
        y = y/np.sqrt(np.trapz(y**2,x)) 
        plt.plot(x,y)
        plt.axis([0,0.23,-2,2])
#rho, R = sol(0,-0.990099, 0.001, N)

#plt.plot(rho,R)
#plt.show
```
