```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#
min_rho = 0.0
max_rho = 20.0
drho=0.01
N= int((max_rho - min_rho)/drho) #cantidad de puntos o datos que se tomaran

#
rho = np.zeros(N)
R = np.zeros(N)
dR = np.zeros(N)
```

```python
eps=-0,99

def ddR(l,eps,rho,R,dR):
    a=-float((2*dR)/rho)
    b=float(((l(l+1))/rho**2) - 2/rho -eps)*R
    return a+b
```

```python
def sol(l,eps,drho,N):
    if l==0:
        R[0]=1
        dR[0]=-1
    else:
        R[0]=0
        dR[0]=1
        
    for i in range (0,N-1):
        rho[i+1] = rho[i] + drho 
        R[i+1] = R[i] + drho * dR[i]
        dR[i+1] = dR[i] + ddR(l,eps,rho[i],R[i],dR[i])*dt 
        
    return rho, R
        
```
