```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#
min_rho = 0.0
max_rho = 20.0
drho=0.01
N= int((max_rho - min_rho)/drho) #cantidad de puntos o datos que se tomaran
```

```python
def ddR(l,eps,rho,R,dR):
    return -float((2*dR)/rho) + float(((l(l+1))/rho**2) - 2/rho -eps)*R

def prob(r, R):
    R = R*R*r*r
    R = R/(np.trapz(R,r))
    return R
```
